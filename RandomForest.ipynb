```python
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
X,_ = make_blobs(centers=5, cluster_std=1, random_state=1)
```

```python
x, y = X[:,0], X[:,1]
```

```python
plt.scatter(x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
mdltr = DecisionTreeRegressor(max_depth=2)
```

```python
xx = x.reshape(-1,1)
```

```python
mdltr.fit(xx,y)
```

```python
yp = mdltr.predict(xx)
```

```python
plt.scatter(x, y)
plt.scatter(x, yp)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score(y,yp)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
modl = RandomForestRegressor(max_depth=3)
```

```python
modl.fit(xx,y)
```

```python
ypp = modl.predict(xx)
```

```python
plt.scatter(x,y)
plt.scatter(xx,ypp)
```
