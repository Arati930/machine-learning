```python
import torch
from torchvision import datasets, transforms
```

```python
dataset = datasets.MNIST(root='./data', train=True, download=True,
                         transform=transforms.ToTensor())
```

```python
x_train = dataset.data
y_train = dataset.targets
```

```python
selected_classes = [0, 1, 2]
```

```python
for cls in selected_classes:
    indices = (y_train == cls).nonzero(as_tuple=True)[0][:100]
    print(indices)
```

```python
x_final, y_final = [], []
for cls in selected_classes:
    indices = (y_train == cls).nonzero(as_tuple=True)[0][:100]
    x_final.append(x_train[indices])
    y_final.append(y_train[indices])
```

```python
len(x_final)
```

```python
xx = torch.cat(x_final)
yy = torch.cat(y_final)
```

```python
yy
```

```python
from torch.utils.data import TensorDataset, DataLoader
```

```python
dataset = TensorDataset(xx, yy)
```

```python
data = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
for batch in data:
  batch_x = batch[0]
  batch_y = batch[1]
  print(batch_x.shape, batch_y.shape)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SimpleNet(nn.Module):
    def __init__(self):
        super(SimpleNet, self).__init__()
        # MNIST images are 28x28 = 784 pixels
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 3)    # 3 output classes

    def forward(self, x):
        # Flatten the image from (28, 28) to (784)
        x = x.view(-1, 784).float()
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model = SimpleNet()
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 10

for epoch in range(epochs):
    running_loss = 0.0
    for images, labels in data:
        # 1. Clear previous gradients
        optimizer.zero_grad()

        # 2. Forward pass: Get predictions
        outputs = model(images)

        # 3. Calculate loss
        loss = criterion(outputs, labels)

        # 4. Backward pass: Calculate gradients
        loss.backward()

        # 5. Optimization: Update weights
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1} - Loss: {running_loss/len(data):.4f}")
```
